##### WAVE PLATE FOR ON STATE
##### WAVE PLATE FOR OFF STATE
##### Jones matrix for 0th order
##### Jones matrix for +1 order

```python
import sympy as smp

theta = smp.pi * smp.Rational(1, 6)#(11, 12)#input wave polarization 30+45+45+45+45= 210 degree  #(11, 12)#input wave polarization 30+45+45+45= 165 degree #(8, 12 )#input wave polarization 30+45+45= 120 degree #(5, 12) #input wave polarization 30+45= 75 degree #(1, 6)  #input wave polarization 30 degree
phi1  = smp.pi * smp.Rational(3, 8) # HWP at 67.5 degree
phi2  = smp.pi * smp.Rational(5, 8) # HWP at 112.5 degree

ji = smp.Matrix([smp.cos(theta), smp.sin(theta)])

Ma = smp.Matrix([
    [smp.cos(2*phi1), smp.sin(2*phi1)],
    [smp.sin(2*phi1), -smp.cos(2*phi1)]
])

Mb = smp.Matrix([
    [smp.cos(2*phi2), smp.sin(2*phi2)],
    [smp.sin(2*phi2), -smp.cos(2*phi2)]
])


M0 = smp.simplify((Ma + Mb)/2)
M0 = smp.trigsimp(M0)
M0 = smp.nsimplify(M0)


j0 = smp.simplify(M0*ji)
j0 = smp.trigsimp(j0) # oth order output

M1 = smp.simplify((Ma - Mb)/2)
M1 = smp.trigsimp(M1)
M1 = smp.nsimplify(M1)

j1 = smp.simplify(M1*ji) # +1 order output
j1 = smp.trigsimp(j1)



angle0 = smp.simplify(j0[1]/j0[0])
angle0 = smp.atan(angle0)
angle0 = smp.trigsimp(angle0)
angle0 = smp.atan2(j0[1], j0[0])
angle0 = smp.nsimplify(angle0, [smp.pi])
angle0

angle1 = smp.simplify(j1[1]/j1[0])
angle1 = smp.atan(angle1)
angle1 = smp.trigsimp(angle1)
angle1 = smp.atan2(j1[1], j1[0])
angle1 = smp.nsimplify(angle1, [smp.pi])
angle0
```

```python
vector0 = smp.latex(j0)
vector1 = smp.latex(j1)
angle0 = smp.latex(angle0)
angle1 = smp.latex(angle1)


with open('output2.txt', 'w') as outf:
    outf.write('input wave at 30 degree  '+ '\n')
    outf.write('j0  =  ' + vector0 + '\n' + 'angle0 = ' + angle0 + '\n')
    outf.write('j1  =  ' + vector1 + '\n' + 'angle1 = ' + angle1 + '\n'+ '\n')
```
